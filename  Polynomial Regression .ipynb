```python
import pandas as pd;
import numpy as np;
from sklearn.linear_model import LinearRegression;
from sklearn.preprocessing import PolynomialFeatures;
import matplotlib.pyplot as plt;


data=pd.read_csv('Position_Salaries.csv')
```

```python
real_x=data.iloc[:,1:2].values
real_y=data.iloc[:,2].values

```

```python
linear_reg=LinearRegression();
linear_reg.fit(real_x,real_y);
```

```python
polynomial_reg=PolynomialFeatures(degree=5);
real_x_poly=polynomial_reg.fit_transform(real_x);

polynomial_reg.fit(real_x_poly,real_y)
linear_reg2=LinearRegression();
linear_reg2.fit(real_x_poly,real_y);
```

```python
plt.scatter(real_x,real_y,color='blue');
plt.plot(real_x,linear_reg.predict(real_x),color='red');
plt.title('Linear Model')
plt.xlabel('postion');
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(real_x,real_y,color='blue');
plt.plot(real_x,linear_reg2.predict(polynomial_reg.fit_transform(real_x)),color='red');
plt.title('Linear Model')
plt.xlabel('postion');
plt.ylabel('salary')
plt.show();
```
